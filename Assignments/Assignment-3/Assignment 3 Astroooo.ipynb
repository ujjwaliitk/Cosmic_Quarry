```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def enhance_image(image_path):
    img = cv2.imread(image_path)

    # Denoise
    denoised = cv2.fastNlMeansDenoisingColored(img, None, 10, 10, 7, 21)

    # Sharpen
    kernel = np.array([[0, -1, 0],
                       [-1, 5, -1],
                       [0, -1, 0]])
    processed_img = cv2.filter2D(denoised, -1, kernel)

    return processed_img

# Image Path
image_path = r"C:\Users\LENOVO\OneDrive\Documents\Mg.jpg"

# Enhance 
enhanced_image = enhance_image(image_path)

# OpenCV reads in BGR format; convert to RGB for matplotlib
enhanced_rgb = cv2.cvtColor(enhanced_image, cv2.COLOR_BGR2RGB)

# Plotting using matplotlib
plt.figure(figsize=(8, 6))
plt.imshow(enhanced_rgb)
plt.title("Processed Image")
plt.axis('off')
plt.show()
```

```python

image_path = r"C:\Users\LENOVO\OneDrive\Documents\K.jpg"
img = cv2.imread(image_path)
if img is None:
    print(" Image not found or cannot be loaded.")
else:

    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    # Calculate intensity for each column (spectrograph)
    def calc_intensity(R, G, B):
        return 0.299 * R + 0.587 * G + 0.114 * B

    height, width, _ = img_rgb.shape
    intensity = []

    for i in range(width):
        column_R = img_rgb[:, i, 0]
        column_G = img_rgb[:, i, 1]
        column_B = img_rgb[:, i, 2]
        column_intensity = calc_intensity(column_R, column_G, column_B)
        intensity.append(np.sum(column_intensity))

    # Generate wavelength scale (400–700 nm)
    wavelength = np.linspace(400, 700, width)

    # Plot spectrograph
    plt.figure(figsize=(10, 4))
    plt.plot(wavelength, intensity, color='blue')
    plt.xlabel("Wavelength (nm)")
    plt.ylabel("Intensity")
    plt.title("Spectrograph of K.png")
    plt.grid(True)
    plt.tight_layout()
    plt.show()



```
