Spectrum of Gold

Au.png

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# ✅ Read uploaded file
file_name = next(iter(uploaded))  # should be au.png now

# Convert uploaded file to OpenCV image
img_array = np.frombuffer(uploaded[file_name], np.uint8)
img = cv2.imdecode(img_array, cv2.IMREAD_COLOR)

# ✅ Denoise the image
denoised = cv2.bilateralFilter(img, 9, 75, 75)

# ✅ Sharpen the image
sharpen_kernel = np.array([[0, -1, 0],
                           [-1, 5, -1],
                           [0, -1, 0]])
sharpened = cv2.filter2D(denoised, -1, sharpen_kernel)

# ✅ Convert to grayscale
gray = cv2.cvtColor(sharpened, cv2.COLOR_BGR2GRAY)

# ✅ Compute spectrum (mean along rows)
spectrum = gray.mean(axis=0)

# ✅ Plot spectrograph
plt.figure(figsize=(10, 4))
plt.plot(spectrum, color='gold')
plt.title("Spectrograph of AU")
plt.xlabel("Pixel Position (Proxy for Wavelength)")
plt.ylabel("Intensity")
plt.grid(True)
plt.show()
```
