```python
df = pd.read_csv("student_age_marks.csv")  # Make sure this file is in the same folder
df.head()
```

```python
plt.figure(figsize=(5, 3))
plt.hist(df['Marks'], bins=10, color='skyblue', edgecolor='black')
plt.title("Histogram of Marks")
plt.xlabel("Marks")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

```python
high_achievers = df[df['Marks'] > 90]
high_achievers
```

```python
plt.figure(figsize=(5, 3))
plt.scatter(df['Age'], df['Marks'], color='purple')
plt.title("Age vs Marks")
plt.xlabel("Age")
plt.ylabel("Marks")
plt.grid(True)
plt.show()
```

```python
df_with_nans = df.copy()
df_with_nans.loc[0, 'Marks'] = np.nan  # Introduce missing value

# Fill missing values with column mean
df_filled = df_with_nans.fillna({'Marks': df_with_nans['Marks'].mean()})
df_filled.head()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file (if not already loaded)
df = pd.read_csv("student_age_marks.csv")

# Create the scatter plot
plt.figure(figsize=(6, 4))
plt.scatter(df['Age'], df['Marks'], color='purple', edgecolor='black', s=80)
plt.title("Age vs Marks")
plt.xlabel("Age")
plt.ylabel("Marks")
plt.grid(True)
plt.tight_layout()
plt.show()
```
