```python
import pandas as pd

df = pd.read_csv('student_age_marks(1).csv')
print(df)
```

```python
import matplotlib.pyplot as plt

plt.hist(df['Marks'], bins=5)
plt.xlabel('Marks')
plt.ylabel('Frequency')
plt.title('Histogram of Frequency vs Marks')
plt.show()
```

```python
filtered_df = df[df['Marks'] > 90]
print(filtered_df)
```

```python
plt.scatter(df['Age'], df['Marks'])
plt.xlabel('Age')
plt.ylabel('Marks')
plt.title('Scatter Plot: Age vs Marks')
plt.show()
```

```python

df = pd.read_csv('student_age_marks(1).csv')
print("Original DataFrame:")
print(df.head())
```

```python

merged_df = pd.merge(df, filtered_df)
print(merged_df)
```

```python
merged_df[['Marks', 'Age']] = merged_df[['Marks', 'Age']].fillna(0)
print(merged_df)
```
