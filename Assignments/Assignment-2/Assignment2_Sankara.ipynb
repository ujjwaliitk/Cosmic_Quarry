```python
import pandas 
import matplotlib.pyplot as plot

df = pandas.read_csv('student_age_marks.csv')

plot.hist(df['Marks'], bins=10, edgecolor='black')

plot.title('Histogram of Students Marks')
plot.xlabel('Marks')
plot.ylabel('Frequency')

# Show the plot
plot.show()
```

```python
import pandas 
import matplotlib.pyplot as plot

df = pandas.read_csv('student_age_marks.csv')

plot.scatter(df['Age'], df['Marks'], color='blue', alpha=0.7)

plot.title('Age vs. Marks')
plot.xlabel('Age')
plot.ylabel('Marks')

plot.show()
```
